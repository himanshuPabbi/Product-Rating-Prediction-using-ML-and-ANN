```python
# cleaning 
chosen_cols = [
    'price', 
    'retail_price',
    'units_sold', 
    'uses_ad_boosts', 
    'badges_count',
    'badge_product_quality',
    'badge_fast_shipping',
    'product_color',
    'product_variation_size_id',
    'origin_country',
    'merchant_id',
    'rating'
]
df_filtered = df[chosen_cols]
```

```python
df_filtered.head().transpose()
```

```python
# encoding 
from sklearn.preprocessing import LabelEncoder

text_features = ['product_color', 'origin_country', 'product_variation_size_id', 'merchant_id']
for feature in text_features:
    if feature in df_filtered.columns:
        df_filtered[feature] = LabelEncoder().fit_transform(df_filtered[feature])
```

```python
df_filtered
```

```python
df_filtered.isna().sum()
```

```python
# visualization 
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,7))
sns.heatmap(df_filtered.corr(), cmap='coolwarm', annot=True)
```

```python
threshold = .005
corr = df_filtered.corr()["rating"]
features_less_than_threshold = corr[abs(corr)<threshold].index

print(f"Dropping features: {features_less_than_threshold}")
df_filtered.drop(columns=features_less_than_threshold, inplace=True, axis=1)
```

```python
df_filtered
```

```python
# splitting 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_filtered.drop('rating', axis=1), df_filtered['rating'], test_size = .2)
```

```python
# train test 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

pipe = Pipeline([
    ("scaler", StandardScaler()), 
    ("regressor", RandomForestRegressor())
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mae= mean_absolute_error(y_test, y_pred)
print(f"Mean absolute error for RandomForestRegressor predicting rating is {mae}")
```

```python
# train test 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

pipe = Pipeline([
    ("scaler", StandardScaler()), 
    ("regressor", DecisionTreeRegressor())
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mae= mean_absolute_error(y_test, y_pred)
print(f"Mean absolute error for DecisionTreeRegressor predicting rating is {mae}")
```

```python
# train test 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error

pipe = Pipeline([
    ("scaler", StandardScaler()), 
    ("regressor", KNeighborsRegressor())
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mae= mean_absolute_error(y_test, y_pred)
print(f"Mean absolute error for KNeighborsRegressor predicting rating is {mae}")
```

```python
# train test 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import mean_absolute_error

pipe = Pipeline([
    ("scaler", StandardScaler()), 
    ("regressor", xgb.XGBRegressor())
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mae= mean_absolute_error(y_test, y_pred)
print(f"Mean absolute error for xgboostregressor predicting rating is {mae}")
```

```python
model=keras.Sequential([
    keras.layers.Dense(15,input_shape=(11,),activation='relu'),
    keras.layers.Dense(1,activation='linear')
])
```

```python
model.compile(optimizer='adam',
    loss='mse',
    metrics=['mse'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=64,validation_split=0.20)
```
